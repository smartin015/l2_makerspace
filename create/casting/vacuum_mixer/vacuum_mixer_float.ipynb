```python
from jupyter_cadquery import set_defaults, set_sidecar
from jupyter_cadquery.cadquery import (PartGroup, Part, show)

set_defaults(theme="dark")
set_sidecar("CadQuery", init=True)
import cadquery as cq

print(f"Imported cadquery version {cq.__version__}")
```

```python
COLUMN_DIA = 50.84 - 2*6.35 # Inner diameter
THICK = 5
COL_H = 20
NCOLS = 10

XY = cq.Workplane("XY")

# A center shaft and an outer cylinder
part = XY.circle(COLUMN_DIA/2 * 0.97).extrude(THICK/2, both=True).faces("|Z").fillet(THICK/2-0.01)

cols = [
    XY.circle(THICK/4)
    .extrude(COL_H)
    .translate((COLUMN_DIA/2 - THICK/2, 0, 0))
    .rotate((0,0,0),(0,0,1), i*360/NCOLS)
    for i in range(NCOLS)]
for c in cols:
    part = part.union(c)
    
part
```

```python
cq.exporters.export(part, "vacuum_mixer_float.stl")
print("done")
```
